```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn import preprocessing
import matplotlib.pylab as plt
import dmba
```

```python
cereals_df = dmba.load_data('Cereals.csv')
cereals_df.head()
```

```python
cereals_df.drop(['name', 'mfr', 'type'], axis=1, inplace=True)
```

```python
cereals_df.head()
```

```python
cereals_df.dropna(inplace=True)
```

```python
pcs = PCA(n_components=2)
pcs.fit(cereals_df)
```

```python
pcsSummary = pd.DataFrame({'Standard deviation': np.sqrt(pcs.explained_variance_),
                           'Proportion of variance': pcs.explained_variance_ratio_,
                           'Cumulative proportion': np.cumsum(pcs.explained_variance_ratio_)})
pcsSummary = pcsSummary.transpose()
pcsSummary.columns = ['PC1', 'PC2']
pcsSummary.round(4)
```

The `components_` field of `pcs` gives the individual components. The columns in this matrix are the principal components `PC1`, `PC2`. The rows are variables in the order they are found in the input matrix, `calories` and `rating`.

```python
pd.DataFrame(pcs.components_.transpose(), columns=['PC1', 'PC2'], index=cereals_df.iloc[:,:].columns)
```
