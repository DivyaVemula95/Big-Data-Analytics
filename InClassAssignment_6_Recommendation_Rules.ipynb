```python
import pandas as pd
import matplotlib.pylab as plt
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
from mlxtend.preprocessing import TransactionEncoder

from surprise import Dataset, Reader, KNNBasic
from surprise.model_selection import train_test_split

from surprise import accuracy, Dataset, SVD
from surprise.model_selection import KFold

import dmba
```

```python
# Load the data set
retail_df = pd.read_csv('Retail.csv', header=None)
retail_df.head(6)
```

```python
# preprocess data set
# preprocess data set
# Check for missing values


# If necessary, fill missing values or drop them
print(retail_df.isnull().sum())
retail_df = retail_df.dropna()
transactions = retail_df.values.tolist()

```

```python
encoder = TransactionEncoder()
onehot = encoder.fit(transactions).transform(transactions)
df = pd.DataFrame(onehot, columns=encoder.columns_)


# Generate frequent item sets with min_support = 0.3
frequent_itemsets = apriori(df, min_support=0.3, use_colnames=True)
```

```python
print(frequent_itemsets)
```

```python
# and convert into rules
rules = association_rules(frequent_itemsets, metric='confidence', min_threshold=0.6)
```

```python
rules
```

```python
rules.sort_values(by=['lift'], ascending=False).head(6)
print(rules.sort_values(by=['lift'], ascending=False)
      .drop(columns=['antecedent support', 'consequent support', 'conviction'])
      .head(6))
```

```python
# filter to get rules with atleast 2 consequents only
rules[[len(c) >=2 for c in rules.consequents]].sort_values(by=['support'], ascending=False)
```

```python
# Load and preprocess data set
ratings_df = pd.read_csv('ratings_small.csv')
ratings_df.head()
```

```python
reader = Reader(rating_scale=(1, 5))
data = Dataset.load_from_df(ratings_df[['userId', 'movieId', 'rating']], reader)
```

```python
sim_options = {'name': 'pearson', 'user_based': True}
algo = KNNBasic(sim_options=sim_options)
```

```python
trainset, testset = train_test_split(data, test_size=0.25, random_state=1)
algo.fit(trainset)
```

```python
predictions = algo.test(testset)
accuracy.rmse(predictions, verbose=True)
```
